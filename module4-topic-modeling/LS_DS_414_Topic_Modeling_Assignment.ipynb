```python
data_path = '/content/Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products_May19.csv'

```

```python
import gensim
from gensim import corpora

from gensim.models.ldamulticore import LdaMulticore
from gensim.models import LdaMulticore
from gensim.corpora import Dictionary

import spacy
import pandas as pd
```

```python
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
```

```python
import spacy.cli
spacy.cli.download("en_core_web_lg")
```

```python
data = pd.read_csv(data_path)
data.head()
```

```python
nlp = spacy.load("en_core_web_lg")
def tokenize(X):
  doc = nlp(X)
  return [token.lemma_ for token in doc if (token.is_stop != True) and (token.is_punct != True)]
```

```python
data['tokens'] = data['reviews.text'].apply(tokenize)
```

```python
data['tokens'].head()
```

```python
id2word = corpora.Dictionary(data['tokens'])
```

```python
id2word.filter_extremes(no_below=5, no_above=0.95)
```

```python
corpus = [id2word.doc2bow(text) for text in data['tokens']]
```

```python
lda = LdaMulticore(corpus=corpus,
                   id2word=id2word,
                   iterations=5,
                   workers=4,
                   num_topics = 9
                  )
```

```python
lda.print_topics()
```

```python

pyLDAvis.gensim.prepare(lda, corpus, id2word)
```
